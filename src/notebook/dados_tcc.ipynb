```python
import pandas as pd
import plotly.express as px

# Derreter o DataFrame para ter uma linha por m√™s por munic√≠pio por ano
# rmvp= rmvp( # <-- Esta linha est√° incorreta e ser√° removida
#     id_vars=['Munic√≠pio de notifica√ß√£o', 'Codigo Munic√≠pio', 'Ano'],
#     value_vars=['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'],
#     var_name='M√™s',
#     value_name='Casos'
# )

rmvp_melted = rmvp.melt(
    id_vars=["Munic√≠pio de notifica√ß√£o", "Codigo Munic√≠pio", "Ano"],
    value_vars=["Jan", "Fev", "Mar", "Abr", "Mai", "Jun", "Jul", "Ago", "Set", "Out", "Nov", "Dez"],
    var_name="M√™s",
    value_name="Casos",
)

# Mapear nomes dos meses para n√∫meros para facilitar a ordena√ß√£o temporal
meses_map = {
    "Jan": 1,
    "Fev": 2,
    "Mar": 3,
    "Abr": 4,
    "Mai": 5,
    "Jun": 6,
    "Jul": 7,
    "Ago": 8,
    "Set": 9,
    "Out": 10,
    "Nov": 11,
    "Dez": 12,
}
rmvp_melted["M√™s_num"] = rmvp_melted["M√™s"].map(meses_map)

# Criar uma coluna de data combinando Ano e M√™s
# Usamos o 28 como dia para evitar problemas com meses que n√£o t√™m 31 dias
rmvp_melted["Data"] = pd.to_datetime(
    rmvp_melted["Ano"].astype(str) + "-" + rmvp_melted["M√™s_num"].astype(str) + "-28"
)

# Converter a coluna 'Casos' para num√©rica (pode haver strings como '-')
rmvp_melted["Casos"] = pd.to_numeric(rmvp_melted["Casos"], errors="coerce").fillna(0).astype(int)

# Ordenar por Data para o gr√°fico de linhas
rmvp_melted = rmvp_melted.sort_values(by="Data")

# Criar o gr√°fico de linhas interativo com Plotly Express
fig = px.line(
    rmvp_melted,
    x="Data",
    y="Casos",
    color="Munic√≠pio de notifica√ß√£o",
    title="Evolu√ß√£o Mensal dos Casos de Dengue por Munic√≠pio no Vale do Para√≠ba (2007-2025)",
)

# Atualizar o layout para garantir que o eixo X mostre o intervalo de datas completo
fig.update_layout(xaxis=dict(rangeslider=dict(visible=True), type="date"))

# Exibir o gr√°fico
fig.show()

# Opcional: verificar o intervalo de datas no dataframe
print("Intervalo de datas no dataframe:")
print(rmvp_melted["Data"].min())
print(rmvp_melted["Data"].max())
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Filtrar os dados para o per√≠odo de 2007 a 2023
rmvp_regressao = rmvp_melted[(rmvp_melted["Ano"] >= "2007") & (rmvp_melted["Ano"] <= "2025")].copy()

# Agrupar por ano e somar o total de casos para a regress√£o anual
rmvp_regressao_anual = rmvp_regressao.groupby("Ano")["Casos"].sum().reset_index()

# Preparar os dados para a regress√£o
# Convertendo o 'Ano' para num√©rico para o modelo
rmvp_regressao_anual["Ano_num"] = pd.to_numeric(rmvp_regressao_anual["Ano"])

# Vari√°vel independente (X): Ano
X = rmvp_regressao_anual[["Ano_num"]]
# Vari√°vel dependente (y): Total de Casos
y = rmvp_regressao_anual["Casos"]

# Criar e treinar o modelo de regress√£o linear
model = LinearRegression()
model.fit(X, y)

# Fazer a previs√£o para 2024
ano_previsao = np.array([[2026]])
previsao_2024 = model.predict(ano_previsao)

# Exibir os resultados da regress√£o
print(f"Coeficiente angular (tend√™ncia anual): {model.coef_[0]:.2f}")
print(f"Intercepto: {model.intercept_:.2f}")
print(f"Previs√£o de casos totais para 2025 (tend√™ncia linear): {previsao_2024[0]:.2f}")

# Opcional: Visualizar os dados e a linha de regress√£o
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color="blue", label="Dados Hist√≥ricos (2007-2025)")
plt.plot(X, model.predict(X), color="red", label="Linha de Regress√£o")
plt.scatter(ano_previsao, previsao_2024, color="green", label="Previs√£o para 2025")
plt.xlabel("Ano")
plt.ylabel("Total de Casos")
plt.title("Regress√£o Linear da Tend√™ncia Anual de Casos de Dengue")
plt.legend()
plt.grid(True)
plt.show()
```
